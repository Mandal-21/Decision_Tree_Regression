```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
dataset = pd.read_csv(r'E:\Udemy\Machine Learning A-Z\Part 2 - Regression\Section 8 - Decision Tree Regression\Position_Salaries.csv')

X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
# Fitting dataset to Decision Tree Regressor

from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X, y)
```

```python
# Prediction

y_pred = regressor.predict([[6.5]])
y_pred
```

```python
# Plotting Graph

X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid)), 1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue', linewidth = 2)
plt.title('Years of Experience vs Salary (Decision Tree Regression)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```
